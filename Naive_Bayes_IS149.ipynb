```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
```

```python
# Convert the `pc` column to integer datatype
df['pc'] = df['pc'].replace({'Low': 0, 'Normal': 1, 'High': 2})
df['pc'] = pd.to_numeric(df['pc'])

# Print the updated dataset
print(df)
```

```python
# Convert the `wbc` column to integer datatype
df['wbc'] = df['wbc'].replace({'Low': 0, 'Normal': 1, 'High': 2})
df['wbc'] = pd.to_numeric(df['wbc'])

# Print the updated dataset
print(df)
```

```python
# Convert the `mc` column to integer datatype
df['mc'] = df['mc'].replace({'Low': 0, 'Normal': 1, 'High': 2})
df['mc'] = pd.to_numeric(df['mc'])

# Print the updated dataset
print(df)
```

```python
# Convert the `ast` column to integer datatype
df['ast'] = df['ast'].replace({'Low': 0, 'Normal': 1, 'High': 2})
df['ast'] = pd.to_numeric(df['ast'])

# Print the updated dataset
print(df)
```

```python
# Convert the `bc` column to integer datatype
df['bc'] = df['bc'].replace({'Low': 0, 'Normal': 1, 'High': 2})
df['bc'] = pd.to_numeric(df['pc'])

# Print the updated dataset
print(df)
```

```python
# Convert the `ldh` column to integer datatype
df['ldh'] = df['ldh'].replace({'Low': 0, 'Normal': 1, 'High': 2})
df['ldh'] = pd.to_numeric(df['ldh'])

# Print the updated dataset
print(df)
```

```python
# Split the dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(df.drop('diagnosis', axis=1), df['diagnosis'], test_size=0.2, random_state=42)
```

```python
# Create a Naive Bayes classifier
clf = GaussianNB()

# Train the classifier
clf.fit(X_train, y_train)
```

```python
# Make predictions on the test set
y_pred = clf.predict(X_test)
```

```python
# Compute the accuracy, precision, recall, and F1 score
accuracy = np.mean(y_pred == y_test)
precision = np.sum(y_pred[y_test == 1] == 1) / np.sum(y_pred == 1)
recall = np.sum(y_pred[y_test == 1] == 1) / np.sum(y_test == 1)
f1_score = 2 * (precision * recall) / (precision + recall)
```

```python
from sklearn import metrics
from sklearn.metrics import roc_curve
from sklearn.metrics import auc
```

```python
# Compute the ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_pred)
roc_auc = auc(fpr, tpr)
```

```python
# Print the results
print('Accuracy:', accuracy)
print('Precision:', precision)
print('Recall:', recall)
print('F1 score:', f1_score)
print('ROC AUC:', roc_auc)
```
